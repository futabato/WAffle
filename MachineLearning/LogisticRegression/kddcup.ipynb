```python
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
```

```python
df_train = pd.read_csv('..\\Dataset\\KDD Cup 1999 Data\\kddcup_train.csv')
df_train.head()
```

```python
print(df_train.shape)
```

```python
null_cols = [col for col in df_train.columns if df_train[col].isnull().sum() > 0]
print(null_cols)
```

```python
df_train[:] = np.nan_to_num(df_train)
```

```python
X_train = df_train.iloc[:, [0, 7, 10, 11, 13, 35, 37, 39]] 
X_train = (X_train - X_train.mean()) / X_train.mean()       
y_train = df_train.iloc[:, [41]]   
```

```python
X_train = X_train.dropna()
y_train = y_train.dropna()
```

```python
df_test = pd.read_csv('..\\Dataset\\KDD Cup 1999 Data\\kddcup_test.csv')
df_test.head()
```

```python
print(df_test.shape)
```

```python
null_cols = [col for col in df_test.columns if df_test[col].isnull().sum() > 0]
print(null_cols)
```

```python
df_test[:] = np.nan_to_num(df_test)
```

```python
X_test = df_test.iloc[:, [0, 7, 10, 11, 13, 35, 37, 39]]
X_test = (X_test - X_test.mean()) / X_test.mean()
y_test = df_test.iloc[:, [41]]
```

```python
X_test = X_test.dropna()
y_test = y_test.dropna()
```

```python
LR =  LogisticRegression(C=1e5)
```

```python
LR.fit(X_train, y_train)
```

```python
probably = LR.predict_proba(X_test)
```

```python
y_pred = LR.predict(X_test)
```

```python
print('score : {0}'.format(metrics.accuracy_score(y_test, y_pred)))
```

```python
idx = 0
for predict, prob in zip(y_pred, probably):
    print('{0}\t{1}\t{2}'.format(y_test.iloc[idx, [0]].values[0], predict, np.max(prob)))
    idx += 1
```
